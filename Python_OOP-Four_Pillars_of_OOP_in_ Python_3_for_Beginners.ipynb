### EXERCISE - ATTRIBUTES AND METHODS
Write an object oriented program to create a precious stone.
Not more than 5 precious stones can be held in possession at a
given point of time. If there are more than 5 precious stones,
delete the first stone and store the new one.

```python
class Stone(object):
    
    def __init__(self, name):
        self.name = name
    
    def __get__(self, instance, owner):
        print('Object {} geted'.format(self.name))
        return(self.name)
    
    def __set__(self, instance, name):
        self.name = name
        print('Object {} seted'.format(self.name))
        
    def __delete__(self, instance):
        print('Object {} deleted'.format(self.name))
        del self.name
        
    def __repr__(self):
        return(self.name)

class Foo(object):
    listStones = []
    keyWords = ['Exit', 'Print']
    objStone = Stone('')
        
    @staticmethod
    def instructionsStone():        
        print("Enter the name of precious stone, {} to list all stones or {} to close\n".format(Foo.keyWords[1][0], Foo.keyWords[0][0]))
        
    @staticmethod
    def listCollection(list):
        for obj in list:
            print(dict(Stone(obj).__dict__))
    
    @staticmethod
    def runProgram():
        inp = ''
        f.instructionsStone()
        
        while inp is f.keyWords[0]:
            inp = input().title()
            
            if inp not in list(f.listStones):
                if inp is f.keyWords[1]:
                    f.listCollection(f.listStones)

                elif inp not in list(f.keyWords[0]):
                    f.objStone = inp
                    f.listStones.append(f.objStone)

                    if len(f.listStones) > 5:
                        f.objStone = getattr(Stone(f.listStones[0]), 'name')
                        del f.objStone
                        f.listStones.remove(f.listStones[0])
                        f.listCollection(f.listStones)
            else:
                print('The precious stone {}\'s already instanced.'.format(inp))
    
f = Foo()
f.runProgram()
```
